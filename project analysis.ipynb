```python
import networkx as nx
import csv
import matplotlib.pyplot as plt
from operator import itemgetter
```

```python
# Read in the nodelist file
with open('/Users/bee/Desktop/fb-pages-food/fb-nodes.csv', 'r') as nodecsv:                 
    nodereader = csv.reader(nodecsv)                                       
    nodes = [n for n in nodereader][1:]                                    

# Get a list of just the node names (the first item in each row)
node_names = [n[1] for n in nodes]                                       

# Read in the edgelist file
with open('/Users/bee/Desktop/fb-pages-food/fb-edges.csv', 'r') as edgecsv:                         
    edgereader = csv.reader(edgecsv)                                   
    edges = [tuple(e) for e in edgereader][1:]                                                                                                       

G = nx.Graph() # Initialize a Graph object                                                        
G.add_nodes_from(node_names) # Add nodes to the Graph                
G.add_edges_from(edges) # Add edges to the Graph
print(nx.info(G))
```

```python
#network visualization
fig = plt.figure(figsize=(17,17)) # It is large so we shold create a large figure space first
nx.draw(G, with_labels=True) #, width=[weight for u,v,weight in G.edges(data='weight')])
plt.savefig('edge_graph.png')
plt.show()
```

```python
for node,degree in G.degree:
    display("Node {} has degree {}".format(node,degree))
```

```python
sorted(G.degree,key=lambda x: x[1], reverse=True)
```

```python
#Histogram of the network
fig = plt.figure(figsize=(5,5)) # stipulates a frame size for the histogram to be shown in
hist_data = nx.degree_histogram(G)   # networkx command to generate histogram data of the nodes' degrees
plt.bar(range(len(hist_data)),hist_data) # using matplotlib to generate the histogram
plt.xlabel('node degree', fontsize=16)
plt.ylabel('number of nodes', fontsize=16)
plt.savefig('histogram_nodes.png')
plt.show()
```

```python
#network density calcuation
density = nx.density(G)
print("Network density is:", density)
```

```python
top_two_degree_path = nx.shortest_path(G, source="265", target="518")

print("Shortest path between node 265 and node 518:", top_two_degree_path)
print("Length of this path:", len(top_two_degree_path)-1)
```

```python
#betweenness
betweenness_dict = nx.betweenness_centrality(G)
nx.set_node_attributes(G, betweenness_dict, 'betweenness')
sorted_betweenness = sorted(betweenness_dict.items(), key=itemgetter(1), reverse=True)

print("Top 20 nodes by betweenness centrality:")
for b in sorted_betweenness[:20]:
    print(b)
```

```python
#closeness
closeness_dict = nx.closeness_centrality(G) # Run betweenness centrality
nx.set_node_attributes(G, closeness_dict, 'closeness')
sorted_closeness = sorted(closeness_dict.items(), key=itemgetter(1), reverse=True)
print("Top 20 nodes by closeness centrality:")
for b in sorted_closeness[:20]:
    print(b)
```

```python
#eigenvector
eigenvector_dict = nx.eigenvector_centrality(G) 
nx.set_node_attributes(G, eigenvector_dict, 'eigenvector')
sorted_eigenvector = sorted(eigenvector_dict.items(), key=itemgetter(1), reverse=True)
print("Top 20 nodes by eigenvector centrality:")
for b in sorted_eigenvector[:20]:
    print(b)
```
